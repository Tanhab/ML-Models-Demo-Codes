```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
data = load_iris()
y = data.target
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df = (df-df.min())/(df.max()-df.min())
```

```python
X_train, X_test, y_train, y_test = train_test_split(df , y, train_size=0.7 ,random_state=51)
```

```python
model = RandomForestClassifier(n_estimators=4, random_state=51)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test,model.predict(X_test))
cm
```

```python
sns.heatmap(cm,annot=True, xticklabels=data.target_names, 
            yticklabels=data.target_names)
```

```python
for i in range(model.n_estimators):
    plot_tree(model.estimators_[i])
    plt.show()
```
